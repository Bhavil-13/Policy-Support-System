```python
import pandas as pd
import numpy as np
import networkx as nx
from collections import defaultdict
import math
```

```python
# df = pd.read_excel('2d_calc.xlsx')
df = pd.read_excel('../input files/2d_calc.xlsx')
```

```python
temp = ['- 2', '- 1', '+ 1', '+ 2']
imr_fh = 'Impact IMR(ANC '
mmr_fh = 'Impact MMR(ANC '
sh = '0%)'
```

```python
def dissonance(df, i, diss_list):
    # abs(imr - mmr)
    for j in range(len(df)):
        dis = abs(df[imr_fh + i + sh][j] - df[mmr_fh + i + sh][j])
        diss_list.append(dis)
```

```python
diss_df = pd.DataFrame()
```

```python
diss_df['Taluka'] = df['Taluka']
```

```python
for i in temp:
    diss_list = []
    dissonance(df, i, diss_list)
    diss_df['DISSONANCE (ANC' + i + '0%)'] = diss_list
```

```python
output = pd.DataFrame()
output['Taluka'] = df['Taluka']
output['Stability(ANC + 10%)'] = df['Stability(ANC + 10%)']
output['Stability(ANC + 20%)'] = df['Stability(ANC + 20%)']
output['Stability(ANC - 10%)'] = df['Stability(ANC - 10%)']
output['Stability(ANC - 20%)'] = df['Stability(ANC - 20%)']

output['Dissonance(ANC + 10%)'] = diss_df['DISSONANCE (ANC+ 10%)']
output['Dissonance(ANC + 20%)'] = diss_df['DISSONANCE (ANC+ 20%)']
output['Dissonance(ANC - 10%)'] = diss_df['DISSONANCE (ANC- 10%)']
output['Dissonance(ANC - 20%)'] = diss_df['DISSONANCE (ANC- 20%)']

output['Impact IMR(ANC + 10%)'] = df['Impact IMR(ANC + 10%)']
output['Impact IMR(ANC + 20%)'] = df['Impact IMR(ANC + 20%)']
output['Impact IMR(ANC - 10%)'] = df['Impact IMR(ANC - 10%)']
output['Impact IMR(ANC - 20%)'] = df['Impact IMR(ANC - 20%)']

output['Impact MMR(ANC + 10%)'] = df['Impact MMR(ANC + 10%)']
output['Impact MMR(ANC + 20%)'] = df['Impact MMR(ANC + 20%)']
output['Impact MMR(ANC - 10%)'] = df['Impact MMR(ANC - 10%)']
output['Impact MMR(ANC - 20%)'] = df['Impact MMR(ANC - 20%)']
output
```

```python
output.to_excel('2d_score_calc.xlsx')
```

```python
def normalize_diss(df, col_str, nd_list):
    for j in range(len(df)):
        norm_dis = 0
        norm_dis = -1 + 2*(df[col_str][j] - min(df[col_str]))/(max(df[col_str]) - min(df[col_str]))
        norm_dis = -1 * norm_dis
        nd_list.append(norm_dis)
```

```python
normalized_dissonance_df = pd.DataFrame()
normalized_dissonance_df['District'] = diss_df['District']
```

```python
for i in temp:
    col_str = 'DISSONANCE (ANC' + i + '0%)'
    nd_list = []
    normalize_diss(diss_df, col_str, nd_list)
    normalized_dissonance_df[col_str] = nd_list
```

```python
normalized_dissonance_df.to_excel('final_normalized_dissonance.xlsx', index=False)
```
