```python
def dissonance_score(df, score_list, i):
    for j in range(len(df)):
        str_imr = 'Impact IMR(ANC ' + i + '0%)'
        str_mmr = 'Impact MMR(ANC ' + i + '0%)'
        str_stress = 'Stability(ANC ' + i + '0%)'

        mean_impact = df[[str_imr, str_mmr]].mean()
        score = (mean_impact[0] * (df[str_stress][j])) / df['Dissonance(ANC ' + i + '0%)'][j]
        score_list.append(score)
```

```python
df_score = pd.DataFrame()
df_score['Taluka'] = df['Taluka']
```

```python
temp = ['+ 2', '+ 1', '- 1', '- 2']
# temp = ['- 1']
for i in temp:
    score_list = []
    dissonance_score(df, score_list, i)
    str = 'Score | ANC ' + i + '0%'
    df_score[str] = score_list
```

```python
df_score.to_excel('zero-to-one_final_2D_score.xlsx')
```

```python
df_3d_diss = pd.read_excel('3d_score/final_3D_normalized_dissonance.xlsx')
df_3d_stability = pd.read_excel('3d_score/Final_Scaled_3D_tableau.xlsx')
```

```python
def dissonance_score_3d(df_diss, df_stability, score_list, i):
    for j in range(len(df_diss)):
        str_imr = 'IMPACT_SCORE_IMR (HM '+ i + '0%)'
        str_mmr = 'IMPACT_SCORE_IMR (HM '+ i + '0%)'
        str_paw = 'IMPACT_SCORE_PAW (HM '+ i + '0%)'
        str_stability = 'STABILITY_SCORE (HM ' + i + '0%)'
        mean_impact = df_stability[[str_imr, str_mmr, str_paw]].mean()
        score = (mean_impact[0] * (df_stability[str_stability][j])) / df_diss['DISSONANCE (HM' + i + '0%)'][j]
        score_list.append(score)
```

```python
df_3D_score = pd.DataFrame()
df_3D_score['District'] = df_3d_diss['District']
```

```python
temp = ['+ 2', '+ 1', '- 1', '- 2']
# temp = ['- 1']
for i in temp:
    score_list = []
    dissonance_score_3d(df_3d_diss, df_3d_stability, score_list, i)
    str = 'Score | HM ' + i + '0%'
    df_3D_score[str] = score_list
```

```python
df_3D_score.to_excel('final_3D_score.xlsx')
```
